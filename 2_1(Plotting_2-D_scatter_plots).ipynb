```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
#Load Iris.csv into a pandas dataFrame.
iris = pd.read_csv("iris.csv")
```

```python
# Q) how many data points do we have??
print(iris.shape) # shape of the matrix or the data table 150 rows and 5 coloumns
```

```python
# Q) what are the five coloumns?
print(iris.columns)
```

```python
#how many data points for each class?
iris["species"].value_counts()
# this is a balanced dataset (number of class labels are same for each label)
```

## 2-D Scatter plot

```python
# the axis: labels and scale
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.show 
# all the plots are in the same color. so we have no idea which ones are setosa, virginica or versicolor
```

```python
# to add color, legend and make it more user friendly,
sns.FacetGrid(iris , hue = "species" , height=5).map(plt.scatter , "sepal_length" , "sepal_width").add_legend()
plt.show()
```

###Observations:
1. Using sepal_length and sepal_width, we can distinguish Setosa species from the others. (linearly separation)
2. Versicolor and Verginica not differentiable clearly
3. 3-D scatter plot will help with this
